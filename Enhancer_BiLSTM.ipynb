```python
import os
from pathlib import Path
import numpy as np
import pandas as  pd
from tqdm import tqdm
from Bio import SeqIO

import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks, utils
```

```python
DATA_DIR = Path("Vista_Dataset")
FASTA_FILE = DATA_DIR / "vista_sequences.fasta"
```

```python
records = list(SeqIO.parse(FASTA_FILE, "fasta"))
data = []
for record in records:
    seq = str(record.seq)
    label = record.description.split()[-1]
    data.append((seq, label))
df = pd.DataFrame(data, columns=["sequence", "label"])
print(df.head())
```
